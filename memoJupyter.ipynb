# Mémo concernant Jupyter NoteBook :

- Le site : https://jupyter.org/


- La documentation : https://jupyter-notebook.readthedocs.io

- Pour tester : https://jupyter.org/try


- https://nbviewer.jupyter.org/ permet de visualiser en l'état (non interactif) tous les Notebooks du web...
> Comme par exemple : https://nbviewer.jupyter.org/github/ericECmorlaix/mesBN/blob/master/index.ipynb
>
> Le bouton <button class='fa fa-download icon-download btn btn-xs btn-default'></button> permet alors de télécharger une copie du fichier en local.


## Pour disposer d'un environnement interactif Jupyter :
### Local :
- La solution la plus pratique est d'installer une distribution Anaconda : https://www.anaconda.com/distribution/
 
### En ligne :

- Chez Microsoft Azure Notebooks : https://notebooks.azure.com/
    > *Le travail de Tony HIRST permet d'explorer : https://notebooks.azure.com/OUsefulInfo/projects/gettingstarted*

- Chez Google Colaboratory : https://colab.research.google.com
    > *Google a participé à l'origine de la création du projet Jupyter puis a poursuivi son développement indépendemment pour des besoins internes avant que de le rendre public en l'ajoutant à sa suite d'outils bureautique. C'est différent mais compatible et permet le partage...*

- Pour ouvrir de façon interactive mais temporaire un dépot GitHub ou GitLab dans un environnement Jupyter : https://mybinder.org/.
    > ***Attention**, il faut ajouter un [fichier de configuration](https://mybinder.readthedocs.io/en/latest/using.html#preparing-a-repository-for-binder) soit en [pip](https://github.com/binder-examples/requirements) soit en [conda](https://github.com/binder-examples/conda) pour indiquer à la machine virtuelle les paquets (Noyaux, Modules, Extensions) à installer.*

- Pour installer son propre serveur JupyterHub :
    - le plus simple, suivre la documentation : https://tljh.jupyter.org/en/latest/
> *[nbgitpuller]() pour distribuer des fichiers supports de cours aux élèves. *

    - autre solution : https://pythonforundergradengineers.com/installing-jupyter-hub.html

## Trucs & astuces :
- pour basculer de l'environnement Jupyter NoteBook classique à l'environnement Lab, il suffit de remplacer dans l'URL le `tree` par un `lab` et inversement...
- pour ne télécharger qu'un seul fichier d'un dépot GitHub (ou GitLab) le visualiser dans [nbviewer](https://nbviewer.jupyter.org/) puis cliquer sur le bouton <button class='fa fa-download icon-download btn btn-xs btn-default'></button>
